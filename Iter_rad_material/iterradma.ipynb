Vamos a simular: una fuente puntual, un material monocomponente. y si el fotón es absorvido o desviado.


> Tenemos una fuente puntual, situada a una distancia d de un material con una altura h un grosor a y un ancho y.
> Vamos a con la fuente lanzar fotones en las 3 direcciones del espacio.
> El fotón avanza hasta encontrarse con un átomo, los átomos tienen un grosor.
> Así habrá fotones que colisionen y fotones que no colisionen.
> Para simular paso del fotón en el material, avanza según su recorrido libre medio:
> p(s)=lambda exp(-lambda S) con lambda = 1/S0. S0 = Nsigma (numero de atomos por unidad de volumen por la sección eficaz).
> sigma = pi r0\*\*2

```python
ate = 0.01

l, I = distribucion(1, 0, 200, ate, 100, modo='otro modo', Nphoto=1000)
ln, In = distribucion(1, 0, 200, ate, 100, Nphoto=1000)

plt.scatter(l, I, label="No normal")
plt.scatter(ln, In, label="incidencia normal")
plt.plot(l,np.exp(-ate*l),'r', label=f'exp(-a*l)')
plt.legend()
plt.title('Distribucion de proporcion de fotones detectados en función del espesor del material')
plt.xlabel('l')
plt.ylabel('I')
plt.show()
```
